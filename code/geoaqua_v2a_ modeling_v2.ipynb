BACK TO PREPROCESSING

```python
#Train Test Split on df1
```

```python
#first define
X = df1.drop(['label'], axis=1)
y = df1['label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=187)
```

```python
# except label, take all columns
cols = X_train.columns.difference(['ID', 'label'])
```

```python
#STANDARDSCALER

sc = StandardScaler()

# normalise only selected columns
X_train[cols] = sc.fit_transform(X_train[cols])
```

```python
print(X_train)
```

PCA on X_train (and save for re-use on X_test)

PCA on X_TRAIN only

```python
cols2 = X_train.columns.difference(['ID'])
```

```python
pca = PCA(n_components=10)
```

```python
components_train = pca.fit_transform(X_train[cols2])
```

```python
#principals = pd.DataFrame(data = components_train
#             , columns = ['pc1', 'pc2','pc3'])

principals = pd.DataFrame(data = components_train
             , columns = ['pc1', 'pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10'])
```

```python
principals.head(1000)
```

```python
print('Explained variability per principal component: {}'.format(pca.explained_variance_ratio_))
```

## MODEL #1:  NOW TRY A LOGISTIC REGRESSION ON THIS SCALED, PCA'd DATA

```python
clf = LogisticRegression(random_state=0).fit(components_train, y_train)
```

```python
y_pred = clf.predict(components_train)
y_pred
```

```python
accuracy = accuracy_score(y_train, y_pred)
print(f"accuracy: {round(accuracy,2)}")
```

```python
clf.score(components_train, y_train)
```

```python
#now run Scaler, PCA on X_test --> then test
```
